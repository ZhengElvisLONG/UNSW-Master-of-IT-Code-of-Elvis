## Question 1: Prime Factorization with Exponents



```python
def prime_factorization(n):
    """
    Computes the prime factorization of a positive integer.
    
    >>> prime_factorization(1)
    []
    >>> prime_factorization(12)
    [(2, 2), (3, 1)]
    >>> prime_factorization(60)
    [(2, 2), (3, 1), (5, 1)]
    >>> prime_factorization(100)
    [(2, 2), (5, 2)]
    >>> prime_factorization(997)
    [(997, 1)]
    >>> prime_factorization(999)
    [(3, 3), (37, 1)]
    """
    prime = []
    result = []
    if n == 1:
        print(result)
        return
    for p in range(2, n + 1):
        ifprime = True
        for i in range(2, p):
            if p % i == 0:
                ifprime = False
        if ifprime:
            prime.append(p)
    for p in prime:
        time = 0
        while n % p == 0:
            time += 1
            n = n / p
        if time:
            result.append((p, time))
    print(result)
```

## Question 2: Subsequence Sum Combinations



```python
def subsequence_sum_combinations(nums, target):
    """
    Finds all subsequences that sum to the target value.
    
    >>> subsequence_sum_combinations([], 5)
    []
    >>> subsequence_sum_combinations([5], 5)
    [[5]]
    >>> subsequence_sum_combinations([2, 4, 6, 8], 8)
    [[2, 6], [8]]
    >>> subsequence_sum_combinations([1, 2, 3, 4, 5], 10)
    [[1, 4, 5], [2, 3, 5], [1, 2, 3, 4]]
    >>> subsequence_sum_combinations([-1, 1, -2, 2, 3, -3], 0)
    [[-1, 1], [-2, 2], [-3, 3], [-1, -2, 3], [-1, 2, -1], [1, -2, 1], [1, 2, -3], [-1, 1, -2, 2], [-1, 1, -3, 3], [-2, 2, -3, 3]]
    """
    result = []
    def backcheck(path, index, current_sum):
        if current_sum == target:
            result.append(path.copy())
            return
        if index >= len(nums):
            return
        backcheck(path + [nums[index]], index + 1, current_sum + nums[index])
        backcheck(path, index + 1, current_sum)
    backcheck([], 0, 0)
    print(result)
```

## Question 4: Polynomial Arithmetic



```python
def polynomial_add(p1, p2):
    """
    Adds two polynomials.
    
    >>> polynomial_add([1, 2, 3], [4, 5, 6])
    [5, 7, 9]
    >>> polynomial_add([1, 2], [3, 4, 5])
    [4, 6, 5]
    >>> polynomial_add([], [1, 2, 3])
    [1, 2, 3]
    >>> polynomial_add([1, 2, 3], [])
    [1, 2, 3]
    >>> polynomial_add([1, -1, 1], [-1, 1, -1])
    [0, 0, 0]
    """
    result = []
    maxlen = max(len(p1), len(p2))
    while len(p1) < maxlen:
        p1.append(0)
    while len(p2) < maxlen:
        p2.append(0)
    for i in range(maxlen):
        result.append(p1[i] + p2[i])
    print(result)

if __name__ == '__main__':
    import doctest
    doctest.run_docstring_examples(polynomial_add, globals(), verbose=True)
```

```python
def polynomial_subtract(p1, p2):
    """
    Subtracts polynomial p2 from p1.
    
    >>> polynomial_subtract([5, 7, 9], [1, 2, 3])
    [4, 5, 6]
    >>> polynomial_subtract([4, 6, 5], [1, 2])
    [3, 4, 5]
    >>> polynomial_subtract([1, 2, 3], [])
    [1, 2, 3]
    >>> polynomial_subtract([], [1, 2, 3])
    [-1, -2, -3]
    >>> polynomial_subtract([1, 2, 3], [1, 2, 3])
    [0, 0, 0]
    """
    result = []
    maxlen = max(len(p1), len(p2))
    while len(p1) < maxlen:
        p1.append(0)
    while len(p2) < maxlen:
        p2.append(0)
    for i in range(maxlen):
        result.append(p1[i] - p2[i])
    print(result)

if __name__ == '__main__':
    import doctest
    doctest.run_docstring_examples(polynomial_subtract, globals(), verbose=True)
```

```python
def polynomial_multiply(p1, p2):
    """
    Multiplies two polynomials.
    
    >>> polynomial_multiply([1, 2], [3, 4])
    [3, 10, 8]
    >>> polynomial_multiply([1, 0, 1], [1, 1])
    [1, 1, 1, 1]
    >>> polynomial_multiply([1, 2, 3], [])
    []
    >>> polynomial_multiply([], [1, 2, 3])
    []
    >>> polynomial_multiply([1, 1], [1, -1])
    [1, 0, -1]
    >>> polynomial_multiply([1, 2, 1], [1, 2, 1])
    [1, 4, 6, 4, 1]
    """
    dic = {}
    result = []
    if not p1 and p2:
        print([])
        return
    for i in range(len(p1) + len(p2) - 1):
        dic[i] = 0
    for i in range(len(p1)):
        for j in range(len(p2)):
            dic[i + j] += p1[i] * p2[j]
    if all(dic[i] == 0 for i in range(len(p1) + len(p2) - 1)):
        print([])
        return
    for i in range(len(p1) + len(p2) - 1):
        result.append(dic[i])
    print(result)
    return
```
