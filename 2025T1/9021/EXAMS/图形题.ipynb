```python
# Question 1: Vertical Snake Letters (Corrected)
# Arranges consecutive letters starting from 'A' in a vertical snake pattern
# in a matrix of size rows x cols.
# The snake starts from the top-left corner, moves down the first column,
# then up the second column, down the third, and so on.
# Letters wrap around from 'Z' to 'A'.
#
# Note that no line has any trailing space.

def vertical_snake_letters(rows, cols):
    '''
    Creates a vertical snake pattern of letters.
    >>> vertical_snake_letters(3, 1)
    A
    B
    C
    >>> vertical_snake_letters(1, 4)
    ABCD
    >>> vertical_snake_letters(3, 4)
    AFGL
    BEHK
    CDIJ
    >>> vertical_snake_letters(4, 3)
    AHI
    BGJ
    CFK
    DEL
    >>> vertical_snake_letters(5, 5)
    AJKTU
    BILSV
    CHMRW
    DGNQX
    EFOPY
    '''
    # INSERT YOUR CODE HERE
    current = 'A'
    matrix = [["" for _ in range(rows)] for _ in range(cols)]
    for i in range(cols):
        for j in range(rows):
            matrix[i][j] = current
            current = chr((ord(current) + 1 - ord('A')) % 26 + ord('A'))
        if i % 2:
            matrix[i].reverse()
    # print(matrix)
    result = list(map(list, zip(*matrix)))
    for row in result:
        for col in row:
            print(col, end = "")
        print("")
    
if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 2: Transposed Spiral Numbers
# Creates a anti-clockwise spiral pattern of consecutive numbers starting from 1
# in a square matrix of size n x n, and then transposes the result.
# The spiral starts at the top-left corner.
#
# Note that no line has any trailing space.

def transposed_spiral_numbers(n):
    '''
    Creates a transposed spiral pattern of numbers.
    n must be a positive integer.
    >>> transposed_spiral_numbers(1)
    1
    >>> transposed_spiral_numbers(2)
    1 4
    2 3
    >>> transposed_spiral_numbers(3)
    1 8 7
    2 9 6
    3 4 5
    >>> transposed_spiral_numbers(4)
    1 12 11 10
    2 13 16 9
    3 14 15 8
    4 5 6 7
    >>> transposed_spiral_numbers(5)
    1 16 15 14 13
    2 17 24 23 12
    3 18 25 22 11
    4 19 20 21 10
    5 6 7 8 9
    '''
    # INSERT YOUR CODE HERE
    left, right, top, bottom = 0, n - 1, 0, n - 1
    current = 1
    result = [["" for _ in range(n)] for _ in range(n)]
    while current <= n ** 2:
        for row in range(top, bottom + 1):
            result[row][left] = str(current)
            current += 1
        left += 1
        for col in range(left, right + 1):
            result[bottom][col] = str(current)
            current += 1
        bottom -= 1
        for row in range(bottom, top - 1, -1):
            result[row][right] = str(current)
            current += 1
        right -= 1
        for col in range(right, left - 1, -1):
            result[top][col] = str(current)
            current += 1
        top += 1
    for row in result:
        print(" ".join(row))

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 3: Columnar Zigzag Letters (Similar to Original Example)
# Arranges a total number of letters (total_letters) starting from 'A'
# into a grid with a specific height (number of rows).
# Letters fill columns sequentially. Odd-numbered columns (1st, 3rd, ...) are filled top-down.
# Even-numbered columns (2nd, 4th, ...) are filled bottom-up.
# If total_letters is not a multiple of height, the last column might be shorter
# and filled in the direction corresponding to its column number.
# The filling stops exactly after total_letters have been placed.
# Letters wrap around from 'Z' to 'A'.
#
# Note that no line has any trailing space.
# Use <BLANKLINE> for empty lines if height > total_letters.

def columnar_zigzag_letters(total_letters, height):
    '''
    Creates a columnar zigzag pattern similar to the original example.
    >>> columnar_zigzag_letters(4, 1)
    ABCD
    >>> columnar_zigzag_letters(3, 5)
    A
    B
    C
    <BLANKLINE>
    <BLANKLINE>
    >>> columnar_zigzag_letters(4, 2)
    AC
    BD
    >>> columnar_zigzag_letters(5, 2)
    ACE
    BD
    >>> columnar_zigzag_letters(6, 2)
    ACE
    BDF
    >>> columnar_zigzag_letters(7, 3)
    ADG
    BE
    CF
    >>> columnar_zigzag_letters(8, 3)
    ADG
    BEH
    CF
    >>> columnar_zigzag_letters(17, 5)
    AFKP
    BGLQ
    CHM
    DIN
    EJO
    >>> columnar_zigzag_letters(27, 4)
    AEIMQUY
    BFJNRVZ
    CGKOSWA
    DHLPTX
    '''
    # INSERT YOUR CODE HERE
    current = 'A'
    num = 1
    result = ["" for _ in range(height)]
    while num <= total_letters:
        for i in range(height):
            result[i] += current
            current = chr((ord(current) + 1 - ord('A')) % 26 + ord('A'))
            num += 1
            if num > total_letters:
                break
    for row in result:
        print(row)


if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 5: Vertical Zigzag Numbers
# Fills a matrix rows x cols with numbers 1, 2, 3... in a vertical zigzag.
# Starts top-left (1), goes down column 1.
# Then diagonally up-right to the top of column 2.
# Then down column 2.
# Then diagonally up-right to the top of column 3, and so on.
#
# Note that no line has any trailing space.

def vertical_zigzag_numbers(rows, cols):
    '''
    Creates a vertical zigzag pattern of numbers.
    >>> vertical_zigzag_numbers(1, 1)
    1
    >>> vertical_zigzag_numbers(3, 1)
    1
    2
    3
    >>> vertical_zigzag_numbers(1, 4)
    1 2 3 4
    >>> vertical_zigzag_numbers(3, 4)
    1 6 7 12
    2 5 8 11
    3 4 9 10
    >>> vertical_zigzag_numbers(4, 3)
    1 8 9
    2 7 10
    3 6 11
    4 5 12
    >>> vertical_zigzag_numbers(5, 5)
    1 10 11 20 21
    2 9 12 19 22
    3 8 13 18 23
    4 7 14 17 24
    5 6 15 16 25
    '''
    # INSERT YOUR CODE HERE
    matrix = [[0 for _ in range(rows)] for _ in range(cols)]
    current = 1
    for i in range(cols):
        for j in range(rows):
            matrix[i][j] = current
            current += 1
        if i % 2:
            matrix[i].reverse()
    result = list(map(list, zip(*matrix)))
    for row in result:
        print(" ".join(map(str, row)))

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 6: Interleaved Vertical Lines
# Fills a matrix of size rows x cols with numbers starting from 1.
# Odd columns (1st, 3rd, ...) are filled top-down with consecutive odd numbers (1, 3, 5...). 
# Even columns (2nd, 4th, ...) are filled top-down with consecutive even numbers (2, 4, 6...). 
# Filling stops when the matrix is full.
#
# Note that no line has any trailing space.

def interleaved_vertical_lines(rows, cols):
    '''
    Creates interleaved vertical lines of odd and even numbers.
    >>> interleaved_vertical_lines(1, 1)
    1
    >>> interleaved_vertical_lines(3, 1)
    1
    3
    5
    >>> interleaved_vertical_lines(1, 4)
    1 2 3 4
    >>> interleaved_vertical_lines(3, 4)
    1 2 3 4
    3 4 5 6
    5 6 7 8
    >>> interleaved_vertical_lines(4, 3)
    1 2 3
    3 4 5
    5 6 7
    7 8 9
    '''
    # INSERT YOUR CODE HERE
    matrix = [[0 for _ in range(cols)] for _ in range(rows)]
    for col in range(cols):
        for row in range(rows):
            matrix[row][col] = 2 * row + 1 + col
    for row in matrix:
        print(" ".join(map(str, row)))

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 7: Hollow Spiral Square
# Creates a hollow spiral pattern using '*' within a square grid of size n x n.
# The spiral starts from the outer boundary and moves inwards, leaving empty spaces.
# Only the 'walls' of the spiral path are drawn.
# n must be a positive integer.
#
# Note that no line has any trailing space.

def hollow_spiral_square(n):
    '''
    Creates a hollow spiral square pattern.
    >>> hollow_spiral_square(1)
    *
    >>> hollow_spiral_square(2)
    **
    **
    >>> hollow_spiral_square(3)
    ***
    * *
    ***
    >>> hollow_spiral_square(4)
    ****
    *  *
    *  *
    ****
    >>> hollow_spiral_square(5)
    *****
    *   *
    * * *
    *   *
    *****
    >>> hollow_spiral_square(6)
    ******
    *    *
    * ** *
    * ** *
    *    *
    ******
    >>> hollow_spiral_square(7)
    *******
    *     *
    * *** *
    * * * *
    * *** *
    *     *
    *******
    '''
    # INSERT YOUR CODE HERE
    matrix = [[" " for _ in range(n)] for _ in range(n)]
    def create(m, n):
        for i in range(m, n):
            matrix[m][i] = '*'
            matrix[n - 1][i] = '*'
            matrix[i][m] = '*'
            matrix[i][n - 1] = '*'
        if n - m <= 3:
            return
        else:
            create(m + 2, n - 2)
    create(0, n)
    for row in matrix:
        print("".join(row))

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 8: Vertical Number Pyramid
# Creates a pyramid shape using numbers, oriented vertically.
# The pyramid has 'height' levels, corresponding to the maximum number reached.
# The numbers increase from 1 up to 'height' and then decrease back to 1.
# Each number is printed on a new line.
#
# Note that no line has any trailing space.

def vertical_number_pyramid(height):
    '''
    Creates a vertical number pyramid.
    >>> vertical_number_pyramid(1)
    1
    >>> vertical_number_pyramid(2)
    1
    2
    1
    >>> vertical_number_pyramid(3)
    1
    2
    3
    2
    1
    >>> vertical_number_pyramid(4)
    1
    2
    3
    4
    3
    2
    1
    >>> vertical_number_pyramid(5)
    1
    2
    3
    4
    5
    4
    3
    2
    1
    '''
    # INSERT YOUR CODE HERE
    result = []
    for i in range(height):
        result.append(i + 1)
    for i in range(height - 1, 0, -1):
        result.append(i)
    for item in result:
        print(item)

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 9: Boustrophedon Letters (Ox-Plowing)
# Arranges consecutive letters starting from 'A' in a boustrophedon (ox-plowing) pattern
# in a matrix of size rows x cols.
# The pattern fills rows sequentially.
# Odd-numbered rows (1st, 3rd, ...) are filled left-to-right.
# Even-numbered rows (2nd, 4th, ...) are filled right-to-left.
# Letters wrap around from 'Z' to 'A'.
#
# Note that no line has any trailing space.

def boustrophedon_letters(rows, cols):
    '''
    Creates a boustrophedon pattern of letters.
    >>> boustrophedon_letters(1, 5)
    ABCDE
    >>> boustrophedon_letters(3, 3)
    ABC
    FED
    GHI
    >>> boustrophedon_letters(4, 4)
    ABCD
    HGFE
    IJKL
    PONM
    >>> boustrophedon_letters(3, 5)
    ABCDE
    JIHGF
    KLMNO
    >>> boustrophedon_letters(5, 2)
    AB
    DC
    EF
    HG
    IJ
    '''
    # INSERT YOUR CODE HERE
    matrix = ["" for _ in range(rows)]
    current = 'A'
    for i in range(rows):
        for _ in range(cols):
            matrix[i] += current
            current = chr((ord(current) + 1 - ord('A')) % 26 + ord('A'))
        if i % 2:
            matrix[i] = matrix[i][::-1]
    for row in matrix:
        print(row)

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# Question 10: Vertical Wave Pattern
# Fills a matrix of size rows x cols with numbers starting from 1.
# Numbers fill columns top-to-bottom for the first column, then bottom-to-top for the second,
# top-to-bottom for the third, and so on, creating a vertical wave.
#
# Note that no line has any trailing space.

def vertical_wave_pattern(rows, cols):
    '''
    Creates a vertical wave pattern of numbers.
    >>> vertical_wave_pattern(1, 1)
    1
    >>> vertical_wave_pattern(3, 1)
    1
    2
    3
    >>> vertical_wave_pattern(1, 4)
    1 2 3 4
    >>> vertical_wave_pattern(3, 4)
    1 6 7 12
    2 5 8 11
    3 4 9 10
    >>> vertical_wave_pattern(4, 3)
    1 8 9
    2 7 10
    3 6 11
    4 5 12
    >>> vertical_wave_pattern(5, 2)
    1 10
    2 9
    3 8
    4 7
    5 6
    '''
    # INSERT YOUR CODE HERE
    matrix = [[0 for _ in range(rows)] for _ in range(cols)]
    current = 1
    for i in range(cols):
        for j in range(rows):
            matrix[i][j] = current
            current += 1
        if i % 2:
            matrix[i].reverse()
    result = list(zip(*matrix))
    for row in result:
        print(" ".join(map(str, row)))

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```

```python
# You can assume that the first two arguments to rectangle() are
# integers at least equal to 0, and that the third argument, if any,
# is a string consisting of an uppercase letter.
#
# The rectangle is read by going down the first column (if it exists),
# up the second column (if it exists), down the third column (if it exists),
# up the fourth column  (if it exists)...
#
# Hint: ord() and chr() are useful.


def rectangle(width, height, starting_from='A'):
    '''
    >>> rectangle(0, 0)
    >>> rectangle(10, 1, 'V')
    VWXYZABCDE
    >>> rectangle(1, 5, 'X')
    X
    Y
    Z
    A
    B
    >>> rectangle(10, 7)
    ANOBCPQDER
    BMPADORCFQ
    CLQZENSBGP
    DKRYFMTAHO
    EJSXGLUZIN
    FITWHKVYJM
    GHUVIJWXKL
    >>> rectangle(12, 4, 'O')
    OVWDELMTUBCJ
    PUXCFKNSVADI
    QTYBGJORWZEH
    RSZAHIPQXYFG
    '''
    # REPLACE PASS ABOVE WITH YOUR CODE
    if width == height == 0:
        return
    matrix = [["" for _ in range(height)] for _ in range(width)]
    current = starting_from
    for i in range(width):
        for j in range(height):
            matrix[i][j] = current
            current = chr((ord(current) + 1 - ord("A")) % 26 + ord("A"))
        if i % 2:
            matrix[i].reverse()
    matrix = list(map(list, zip(*matrix)))
    for i in range(height):
        print("".join(matrix[i]))
        

if __name__ == '__main__':
    import doctest
    doctest.testmod()
```
